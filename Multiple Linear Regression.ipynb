# Multiple Linear Regression

a statistical technique that uses several independent variables to predict the outcome of a dependent variable

## Objectives

*   Use scikit-learn to implement Multiple Linear Regression
*   Create a model, train it, test it and use the model

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
!wget -O FuelConsumption.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/FuelConsumptionCo2.csv
```

```python
df = pd.read_csv("FuelConsumption.csv")
# take a look at the dataset
df.head()
```

```python
msk = np.random.rand(len(df)) < 0.8
train = df[msk]
test = df[~msk]
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
x = np.asanyarray(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(test[['CO2EMISSIONS']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```

### What is the better group of independent variables

```python
a = np.asanyarray(train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB_MPG']])
b = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (a, b)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

```python
y_hats= regr.predict(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB_MPG']])
c = np.asanyarray(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB_MPG']])
d = np.asanyarray(test[['CO2EMISSIONS']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hats - d) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(c, d))
```

### Conclusion 

ENGINESIZE,CYLINDERS,FUELCONSUMPTION_COMB_MPG - model has a higher variance of 0.87 and closer to perfect score of 1
